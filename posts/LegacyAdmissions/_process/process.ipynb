```python
import pandas as pd
```

```python
data = pd.read_csv('Data_7-22-2025---215.csv')
data = data.iloc[:, :-1]
data = data.dropna()
data.columns = ['id', 'name', 'sector', 'CarnegieClass', 'state', 'lat', 'lon', 'legacy_status', 'enrollment', 'tuition']
data['legacy_status'] = data.legacy_status.apply(int)
data['enrollment'] = data.enrollment.apply(int)
data['tuition'] = data.tuition.apply(int)
data = data[data.CarnegieClass > 0]
data.head()
```

```python
data.to_csv('college_legacy.csv', index=False)
```
